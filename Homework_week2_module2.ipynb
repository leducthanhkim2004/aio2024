```python
import numpy as np

# Giả sử 'data' là mảng NumPy của bạn với cột 'Sales' ở chỉ số 3
Sales = data[:, 3]

# Tính giá trị trung bình của cột 'Sales'
mean_sales = np.mean(Sales)

# Tìm giá trị gần nhất với giá trị trung bình
nearest_value = Sales[np.argmin(np.abs(Sales - mean_sales))]

# Tạo mảng mới 'scores'
scores = np.empty(Sales.shape, dtype=object)

# Gán giá trị cho mảng 'scores'
scores[Sales > nearest_value] = 'Good'
scores[Sales < nearest_value] = 'Bad'
scores[Sales == nearest_value] = 'Average'

# In kết quả scores[7:10]
print(scores[7:10])
```
