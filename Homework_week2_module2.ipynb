```python
import numpy as np

# Gi s 'data' l mng NumPy ca bn vi ct 'Sales'  ch s 3
Sales = data[:, 3]

# Tnh gi tr trung bnh ca ct 'Sales'
mean_sales = np.mean(Sales)

# Tm gi tr gn nht vi gi tr trung bnh
nearest_value = Sales[np.argmin(np.abs(Sales - mean_sales))]

# To mng mi 'scores'
scores = np.empty(Sales.shape, dtype=object)

# Gn gi tr cho mng 'scores'
scores[Sales > nearest_value] = 'Good'
scores[Sales < nearest_value] = 'Bad'
scores[Sales == nearest_value] = 'Average'

# In kt qu scores[7:10]
print(scores[7:10])
```
